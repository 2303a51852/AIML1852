```python
#PART 2

import matplotlib.pyplot as plt

# Dataset
people = ['Kiran', 'Arun', 'Vijay', 'Varun']
age = [25, 30, 35, 40, 45]
height = [145, 151, 165, 173]
weight = [45, 55, 65, 75]

# 3. Scatter Plot: Age vs Height
plt.figure(figsize=(8, 6))
plt.scatter(age[:4], height, color='red')
plt.title('Scatter Plot: Age vs Height')
plt.xlabel('Age (years)')
plt.ylabel('Height (cm)')
plt.grid(True)
plt.show()

# 4. Bar Chart: People vs Weight
plt.figure(figsize=(8, 6))
plt.bar(people, weight, color='lightblue')
plt.title('Bar Chart: People vs Weight')
plt.xlabel('People')
plt.ylabel('Weight (kg)')
plt.show()

# 5. Histogram: Distribution of Heights
plt.figure(figsize=(8, 6))
plt.hist(height, bins=5, color='beige', edgecolor='black')
plt.title('Histogram: Distribution of Heights')
plt.xlabel('Height (cm)')
plt.ylabel('Frequency')
plt.show()
```
